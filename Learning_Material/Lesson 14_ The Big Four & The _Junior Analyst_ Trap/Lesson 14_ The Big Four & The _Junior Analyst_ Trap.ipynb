### âš ï¸ The Engineering Reality: "Confidence != Truth"

#### 1. The Knowledge Cutoff
* **The Problem:** Models are frozen in time. If you ask Gemini to fix a library released *yesterday*, it will confidently Hallucinate a fix using APIs that don't exist, or lecture you that the library doesn't exist.
* **The Fix:** **RAG (Retrieval Augmented Generation)**. We must inject current documentation into the context window (we will learn this in Module 3).

#### 2. The "Junior Analyst" Metaphor
* **Concept:** Treat LLMs like a brilliant intern who has read every book in the world but has zero real-world experience.
* **The Danger:** If a Junior Analyst hits a roadblock (e.g., a "Model Not Found" error), they might panic and write 400 lines of code to patch the library instead of checking if the model name was just a typo.
* **Your Role:** You are the **Senior Engineer**. You must review the strategy, not just the output.
    * *Bad:* "Write this code for me." -> (Copy/Paste blindly).
    * *Good:* "Propose a solution." -> (Review architecture) -> "Okay, implement it."

#### 3. Stack Overflow's Decline
* **Observation:** Developers rarely search for error codes anymore; they paste them into Claude/ChatGPT.
* **Risk:** This creates a feedback loop where we stop understanding the *root cause* because the AI fixes the *symptom* too quickly.

**"Trust, but Verify."**
When generating code, always ask yourself: *"Did it just solve the problem, or did it write a workaround for a problem that shouldn't exist?"*
