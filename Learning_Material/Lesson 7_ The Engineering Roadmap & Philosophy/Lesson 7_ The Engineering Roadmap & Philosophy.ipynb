The course is structured around three distinct layers of abstraction. As engineers, we must master all three:

#### 1. Models (The Foundation)
* **Open Source vs. Closed Source:** When to use Llama 3 vs. GPT-4.
* **Multimodal:** Handling Images and Audio.
* **Selection Strategy:** Choosing the right model for the specific task (Latency vs. Intelligence).

#### 2. Tools & Frameworks (The Glue)
* **Hugging Face:** The hub for models and datasets.
* **LangChain:** Orchestration of chains.
* **Gradio:** Rapid UI prototyping (similar to NiceGUI but for ML).
* **Weights & Biases:** Experiment tracking.

#### 3. Techniques (The Logic)
* **Prompt Engineering:** Few-shot, Chain-of-Thought.
* **RAG (Retrieval Augmented Generation):** Connecting to external data.
* **Fine-Tuning:** Specializing models on custom data.
* **Agents:** Autonomous systems that use tools.
