# Module 2: Frontier Models
## Lesson 8: The Model Landscape (Closed vs. Open)

### üìÑ Overview
This lesson defines the "Frontier Model" landscape. We categorize models into two primary buckets: **Closed Source** (Proprietary/Paid) and **Open Source** (Open Weights/Free to Run). The "Frontier" typically refers to the cutting-edge closed models that require massive capital to train.



### üóùÔ∏è Key Concepts

#### 1. Closed Source (The "Frontier")
* **Definition:** Models where the weights, training data, and architecture are proprietary. You access them via API (Pay-per-token).
* **The "Big Four" Labs:**
    * **OpenAI (GPT Series):** The market leader. Known for **GPT-4o** (and the instructor's future-referenced "GPT-5"). Best general-purpose reasoning.
    * **Anthropic (Claude Series):** Founded by former OpenAI employees. Known for **Claude 3.5 Sonnet** (often preferred by coders). Focuses on safety and massive context windows.
    * **Google (Gemini Series):** Deep integration with the Google ecosystem. **Gemini 1.5 Pro** has a massive context window (up to 2M tokens).
    * **xAI (Grok):** Elon Musk's lab. Known for "Grok," which has real-time access to X (Twitter) data.

#### 2. Open Source (Open Weights)
* **Definition:** Models where the parameters (weights) are public. You can run them locally (Ollama) or host them on your own cloud.
* **Nuance:** Often called "Open Weights" because while you get the model, you don't always get the training data or the recipe used to create it.
* **Key Players:** Meta (Llama), Mistral, and specialized community fine-tunes.

### üß† Engineering Takeaway
As an LLM Engineer, you do not "marry" one model. You choose the right tool for the job.
* *Need Logic/Math?* Use **OpenAI o1**.
* *Need Coding?* Use **Claude 3.5 Sonnet**.
* *Need to process a 500-page PDF?* Use **Gemini 1.5 Pro**.
